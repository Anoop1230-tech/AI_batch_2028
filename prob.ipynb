```python
import random
def cal_p(A,B):
    count=0
    for i in range(0,1000):
        A=0.5
        B=0.4
        if (random.uniform(0.0, 1.0))>A:
            A=1
        else:
            A=0
        if (random.uniform(0.0, 1.0))>B:
            B=0
        else:
            B=1  
        if A*B==1:
            count+=1
    return count
print(cal_p(0.5,0.4))      

    

```

```python
a=0
for i in range(0,10):
    print(cal_p(0.5,0.4))
    a+=(cal_p(0.5,0.4))
print("we found :" ,a//10)

```

```python
results = []
a=0
for i in range(110):  # 50 iterations to see fluctuations
    results.append(cal_p(0.5, 0.4))
    a+=(cal_p(0.5, 0.4))
print(results)
print("Average over 110 iterations:", a // 110)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(results, marker='o', linestyle='-', color='blue')
plt.title("Fluctuation of cal_p over Iterations")
plt.xlabel("Iteration")
plt.ylabel("cal_p Output")
plt.grid(True)
plt.show()
```

```python
plt.plot(results, marker='o', linestyle='-', color='blue', label='Fluctuating Values')
plt.plot( a // 110, linestyle='-', color='red', label='Convergence (Running Average)')
plt.title("Fluctuation and Convergence of cal_p")
plt.xlabel("Iteration")
plt.ylabel("cal_p Output")
plt.legend()
plt.grid(True)
plt.show()
```
